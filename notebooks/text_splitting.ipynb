**Разделение текста на чанки:**

```python
def simple_text_splitter(text: str, chunk_length: int = 200) -> list[str]:
    return [text[i:i+chunk_length] for i in range(0, len(text), chunk_length)]
```

```python
N = 10

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=20,
    length_function=len,
    is_separator_regex=False,
)

chunks = text_splitter.create_documents([text])

print(f"Всего чанков: {len(chunks)}")
print(f"Первые {N} чанков:")
chunks[:N]
```

**Создание цепочки для подготовки текста:**

```python
prompt = PromptTemplate(
    template=open("prompt_split.txt", "r").read(),
    input_variables=['text']
)

model = GigaChat(
    credentials=GIGACHAT_AUTH_KEY,
    scope=GIGACHAT_SCOPE,
    model=GIGACHAT_MODEL_NAME,
    verify_ssl_certs=False,
    profanity_check=False
)

parser = StrOutputParser()
```

```python
chain = prompt | model | parser
```
